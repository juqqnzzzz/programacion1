```python
def codigo(m):
  if isinstance(m, int) and m >= 0:
    for n in range(0, (m+1)):
      if n == 1 or n == 0:
        print("El número es ", n)
      else:
        for x in range(2, n):
          if n % x == 0:
            print(n, 'es igual a', x, '*', n // x)
            break
        else: # Este else corresponde al for, no al if
          print(n, 'es un número primo')
  else:
    print("El número no es natural")
```

```python
codigo(5)
codigo(-3)
codigo("a")
codigo(3.8)
```
