```python
palabra = str(input("ingresa una palabra: "))
vocales = "aeiou"
consonante = "bcdfghjklmnñpqrstvwxyz"
contador = 0
contador1 = 0
for letra in palabra.lower():
  if letra in vocales:
    contador += 1
  elif letra in consonante:
    contador1 += 1
  else:
    continue
print("Cantidad de vocales:", contador)
print("Cantidad de consonantes", contador1)
```
